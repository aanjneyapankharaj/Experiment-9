```python
# Creating a DataFrame
data={
    "Name":["A","B","D"],
    "Marks":[78,90,88]
}
df=pd.DataFrame(data)
df
```

```python
# Access Specific rows
df.loc[1]
```

```python
# Add a New Column
df["Grade"]=["First Class","Distinction","Second Class"]
df
```

```python
# Update Data in DataFrame
# Change marks of student A to 88
df.loc[0,"Marks"]=88
df
```

```python
# Delete a column
df.drop("Grade",axis=1,inplace=True)
df
```

```python
# Perform Basic Statistical Analysis
mean1=df["Marks"].mean()
mean1
```

```python
# To find the minimum and maximum value of marks
min1=df["Marks"].min()
max1=df["Marks"].max()
print(min1,max1)
```

```python
# Apply a condition (Filtering)
# Display students scoring more than 89
df[df["Marks"]>89]
```
